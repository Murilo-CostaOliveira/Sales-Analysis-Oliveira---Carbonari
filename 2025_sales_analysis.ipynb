```python
clientes = []
cliente_atual = None

for _, row in df_vendas.iterrows():
    valor = str(row['Unnamed: 0'])
    if valor.startswith('Cliente:'):
        cliente_atual = valor.replace('Cliente:', '').strip()
    elif pd.notna(row['Unnamed: 6']) and pd.notna(row['Unnamed: 7']):
        if 'Totais' not in valor and 'Relatório' not in valor:
            clientes.append({
                'cliente': cliente_atual,
                'produto': valor,
                'quantidade': row['Unnamed: 6'],
                'valor': row['Unnamed: 7']
            })

df_detalhado = pd.DataFrame(clientes)
print(df_detalhado.head(10))
print(df_detalhado.shape)
```
